```python
import pandas as pd 

mahasiswa = {'tinggi' : [160, 162, 166, 170, 172, 176], 
             'berat' : [56, 58, 60, 64, 70, 72]}

mahasiswa_df = pd.DataFrame(mahasiswa)
mahasiswa_df
```

```python
import matplotlib.pyplot as plt 

mahasiswa_df.plot (kind='scatter', x='berat', y='tinggi')

plt.title('Perbandingan Tinggi dan Berat Mahasiswa')
plt.xlabel('Berat (kg)')
plt.ylabel('Tinggi (cm)')
plt.xlim (55,80)
plt.ylim (155,180)
plt.grid(True)
plt.show
```

```python
import numpy as np 

x = np.array(mahasiswa_df['tinggi'])
y = np.array(mahasiswa_df['berat'])

print(f'x : {x}')
print(f'y : {y}')

```

```python
x = x.reshape(-1, 1)
x.shape
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
```

```python
x_vis = np.array([55,80]).reshape(-1,1)
y_vis = model.predict(x_vis)
```

```python
plt.scatter(x,y)
plt.plot(x_vis, y_vis, '-r')

plt.title('Perbandingan Tinggi dan Berat Mahasiswa')
plt.xlabel('Berat (kg)')
plt.ylabel('Tinggi (cm)')
plt.xlim (55,80)
plt.ylim (155,180)
plt.grid(True)
plt.show
```

```python
print(f'intercept : {model.intercept_}')
print(f'slope : {model.coef_}')
```

```python
print (f'x:\n{x}\n')
print(f'x flatten: {x.flatten()}\n')
print(f'y: {y}')
```

```python
variance_x = np.var(x.flatten(), ddof=1)
print(f'variance: {variance_x}')
```

```python
np.cov(x.flatten(), y)
```

```python
covariance_xy = np.cov(x.flatten(), y)[0][1]

print(f'covariance: {covariance_xy}')
```

```python
slope = covariance_xy / variance_x

print(f'slope: {slope}')
```

```python
intercept = np.mean(y) - slope * np.mean(x)

print(f'intercept: {intercept}')
```

```python
import numpy as np 

x_train = np.array([160, 162, 166, 170, 172, 176]).reshape(-1, 1)
y_train = np.array([56, 58, 60, 64, 70, 72])

x_test = np.array([161, 164, 168, 171, 174, 178]).reshape(-1,1)
y_test = np.array([57, 61, 65, 69, 73, 77])
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score 
y_pred = model.predict(x_test)
r2 = r2_score(y_test, y_pred)
print(f'R-squeared: {r2}')
```
